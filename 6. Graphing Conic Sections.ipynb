### Graphing Conic Sections


#### 1. Graphing parabolas

- General Formula: $$4p(y - k) = (x - h)^2$$

- Simplified for $y$: $$y = \frac{1}{4p}(x-h)^2 + k$$

- Where: 
    - $h$ is shift on the $x$ axis 
    - $k$ is shift on the $y$ axis

```python
# Parameters
a = 1  # 1/4y
h = 1
k = -2
n = 100

# x-axis points to evaluate the function
x = np.linspace(h-5, h+5, n)

# Create the function
y = a*(x-h)**2 + k

plt.plot(x, y)
plt.grid()

plt.show()
```

#### 3. Graphing Circles

##### 3.1 Circle Formula: $$(y - k)^2 + (x - h)^2 = r^2$$

- Or (when the circle is centered at the origin $(0, 0)$): $$x^2 + y^2 = r^2$$


- Where:

    - $r$ is the radius
    - $(h, k)$ are the coordinates of the center on the $x$ and $y$ axis

```python
h = 2
k = -3
r = 3
axlim = r + np.max((abs(h), abs(k)))

x = np.linspace(-axlim, axlim, 100)

X, Y = np.meshgrid(x, x)

Fxy = (X-h)**2 + (Y - k)**2 - r**2

plt.contour(X, Y, Fxy, 0)  # 0 plots only the first contour
plt.plot([-axlim, axlim], [0, 0], 'k--')
plt.plot([0, 0], [-axlim, axlim], 'k--')
plt.plot(h, k, 'go')

plt.gca().set_aspect('equal')

plt.show()
```

#### 4. Graphing Ellipses

##### 4.1  Ellipse Formula: 

$$ \frac{(y - k)^2}{b^2} + \frac{(x - h)^2}{a^2} - 1 = 0$$

- Or (when the ellipse is centered at the origin $(0, 0)$): $$\frac{x^2}{a^2} + \frac{y^2}{b^2} - 1 = 0$$


- Where:

    - $a, b$ are the radii, or the distance from the center on the $x$ and $y$ axes
    - $(h, k)$ are the center coordinates on the $x$ and $y$ axes

```python
h = 1
k = 2
a = 2
b = 3

axlim = np.max((a, b)) + np.max((abs(h), abs(k)))
x = np.linspace(-axlim, axlim, 100)

X, Y = np.meshgrid(x, x)

# Create the function
ellipse = (X-h)**2 / a**2 + (Y-k)**2 / b**2 - 1

plt.plot([-5, 5], [0, 0], 'k--', alpha=0.5)
plt.plot([0, 0], [-5, 5], 'k--', alpha=0.5)
plt.contour(X, Y, ellipse, 0)
plt.plot(h, k, 'go')

plt.gca().set_aspect('equal')
plt.grid()

plt.show()
```

#### 5. Graphing Hyperbolas

##### 5.1  Hyperbola Formula: 

$$ \frac{(y - k)^2}{b^2} - \frac{(x - h)^2}{a^2} - 1 = 0$$

- Or (when the hyperbola is centered at the origin $(0, 0)$): $$\frac{x^2}{a^2} - \frac{y^2}{b^2} - 1 = 0$$


- Where:

    - $a, b$ are the radii, or the distance from the center on the $x$ and $y$ axes
    - $(h, k)$ are the center coordinates on the $x$ and $y$ axes

```python
a = 1
b = 0.5
h = 1
k = 2

axlim = 2*np.max((a, b)) + np.max((abs(h), abs(k)))

x = np.linspace(-axlim, axlim, 100)
X, Y = np.meshgrid(x, x)

# Create function = (X-h)**2 / a**2 + (Y-k)**2 / b**2 - 1
hyperbola = ellipse = (X-h)**2 / a**2 - (Y-k)**2 / b**2 - 1

plt.contour(X, Y, hyperbola, 0)
plt.plot([-axlim, axlim], [0, 0], 'k--', alpha=0.2)
plt.plot([0, 0], [-axlim, axlim], 'k--', alpha=0.2)
plt.plot(h, k, 'go')

plt.gca().set_aspect('equal')

plt.show()
```
