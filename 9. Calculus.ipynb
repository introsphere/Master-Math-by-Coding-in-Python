#### 1. Computing Limits of a Function

- The limit of a function is the value that the function approaches as the input approaches a particular point

- Notation: $ \lim\limits_{x \to a} f(x) $
    
    - Read as: The limits of $f(x)$ as $x$ approaches $a$
    
<br>

- Example: $ \lim\limits_{x \to 4} \frac{x^2}{2} = \frac{4^2}{2} = 8$

```python
x = sp.symbols('x')

fx = x**3
lim_pnt = 1.5
lim = sp.limit(fx, x, lim_pnt)

Math('\\lim_{x \\to %g} %s = %g' %(lim_pnt, sp.latex(fx), lim))
```

```python
fxx = sp.lambdify(x, fx)

xx = np.linspace(-5, 5, 200)

plt.plot(xx, fxx(xx), label='$f(x) = x^3$')
plt.plot(lim_pnt, lim, 'ro', label=f'$x = {lim_pnt}$')

plt.legend()

plt.show()
```

```python
fx = (x**2) / (x-2)
fxx = sp.lambdify(x, fx)
xx = np.linspace(1, 3, 1001)

lim_pnt = 2
lim = sp.limit(fx, x, lim_pnt, dir='-')
plt.plot(xx, fxx(xx))

display(Math('\\lim_{x \\to %g^-} %s = %g' %(lim_pnt, sp.latex(fx), lim)))

plt.show()
```

#### 2. Piecewise Functions

$$f(x) = \begin{cases}
  0, & \text{if } x \leq 0 \\
  -2x, & \text{if } x > 0
\end{cases} $$

- A function that cannot be expressed, using one simple equation

```python
from matplotlib.widgets import Slider
%matplotlib widgets 
# Create the figure and axis
fig, ax = plt.subplots()
plt.subplots_adjust(bottom=0.25)  # Adjust for the slider space

# Create initial data for the plot
t = np.linspace(0, 10, 1000)
initial_frequency = 1.0
y = np.sin(initial_frequency * t)

# Plot the sine wave
line, = ax.plot(t, y)

# Add a slider to control frequency
axcolor = 'lightgoldenrodyellow'
ax_freq = plt.axes([0.1, 0.01, 0.65, 0.03], facecolor=axcolor)
slider = Slider(ax_freq, 'Frequency', 0.1, 10.0, valinit=initial_frequency)

# Define an update function to change the plot based on slider
def update(val):
    freq = slider.val
    line.set_ydata(np.sin(freq * t))
    fig.canvas.draw_idle()  # Redraw the canvas to reflect changes

slider.on_changed(update)

# Display the plot with the widget
plt.show()
```
