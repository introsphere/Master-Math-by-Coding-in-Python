### Graphing and Visualization

#### 1. Plotting coordinates on a plane

```python
x = 3
y = 5

plt.plot(x, y, 'ro')

plt.axis('square')
plt.axis([-6, 6, -6, 6])
plt.grid()

plt.show()
```

```python
x = [-3, 5, -2, 1, 6, 4]
y = [2, 4, -4, 3, -5, 1]

for i in range(0, len(x)):
    plt.plot(x[i], y[i], 'o', label='point %s' %(1+i))

plt.axis('square')
plt.grid()
plt.legend()

plt.show()
```

```python
plt.plot(4, 2, 'r+')

axis = plt.gca()
ylim = axis.get_ylim()
print(ylim)
axis.set_ylim([ylim[0], 5])

plt.show()
```

##### 2.2 Slope-Intercept Form

```python
# y = mx + b

x = np.array([-5, 5])
m = 2
b = 1

y = m*x + b

plt.plot(x, y, label='y = %sx +%s' %(m, b))
plt.axis('square')
plt.xlim(x)
plt.ylim(x)
plt.grid()

axis = plt.gca()
plt.plot(axis.get_xlim(), [0, 0], 'k--')
plt.plot([0, 0], axis.get_ylim(), 'k--')

plt.legend()

plt.show()
```

#### 3. Graphing Rational Functions

- Comes from the word ratio

    - Essentially a ratio of polynomials
    
    - i.e. $y = 2 - x^2$  

```python
x = range(-3, 4)
y = np.zeros(len(x))

x, y
```

```python
for i in range(0, len(x)):
    y[i] = 2 - x[i]**2

plt.plot(x, y, 's-')

plt.show()
```

```python
x = np.linspace(-3, 4, 100)
y = 2 + np.sqrt(abs(x))

plt.plot(x, y, 'm-')

plt.show()
```

#### 4. Plotting with SymPy

```python
x = sp.symbols('x')

y = x**2

p = symplot(y, show=False)

p.xlim = [0, 50]
p[0].line_color = 'm'
p.title = '$x^2$'

p.show()
```

```python
x, a = sp.symbols('x, a')

expr = a/x
n = 1

p = symplot(expr.subs(a, 1), (x, -5, 5), show=False)
p[0].label = '$y = %s$' %sp.latex(expr.subs(a, n))

p.extend(symplot(expr.subs(a, 2), (x, -5, 5), show=False))
p[1].label = '$y = %s$' %sp.latex(expr.subs(a, 2))

p[0].line_color = 'r'
p.ylim = [-5, 5]
p.legend = True

p.show()
```

##### Exercise

```python
x, a = sp.symbols('x, a')

y = a/(x**2-a)

p = symplot(y.subs(a, 1), (x, -5, 5), show=False)
p[0].label = '$%s$' %sp.latex(y.subs(a, 1))


for ia in range(2, 5):
    p.extend(symplot(y.subs(a, ia), (x, -5, 5), show=False))
    p[ia-1].label = '$%s$' %sp.latex(y.subs(a, ia))
    p[ia-1].line_color = np.random.rand(3)

p.ylim = [-10, 10]

p.legend = True
p.show()    
```

#### 5. Making images from matrices

```python
A = [[1, 2], [1, 4]]

plt.imshow(A)

plt.xticks([0, 1])
plt.yticks([0, 1])

plt.show()
```

```python
A = np.zeros((10, 14))
A.shape
```

```python
for i in range(0, A.shape[0]):
    for j in range(0, A.shape[1]):
        A[i, j] = 3*i - 4*j

# plt.plot([0, 3], [8, 2], 'r', linewidth=4)

for i in range(0, A.shape[0]):
    for j in range(0, A.shape[1]):
        plt.text(j, i, A[i, j].astype(int), horizontalalignment='center', verticalalignment='center')

plt.imshow(A)
plt.set_cmap('twilight_shifted')

plt.show()
```

#### 6. Drawing Patches with Polygons

```python
y = np.array([[1, 1], [2, 3], [3, 1]])
y1 = np.array([[2, 2], [2.5, 4], [3.5, 1]])

p = Polygon(y, facecolor='m', alpha=0.5, edgecolor='k')
p1 = Polygon(y1, facecolor='b', alpha=0.5, edgecolor='k')

fig, ax = plt.subplots()

ax.add_patch(p)
ax.add_patch(p1)


ax.set_ylim([0, 4])
ax.set_xlim([0, 4])


plt.show()
```

```python
x = np.linspace(-2, 2, 101)
fx = -x**2

y = np.vstack((x, fx)).T
y1 = np.array([[-0.5, -4], [-0.5, -2.5], [0.5, -2.5], [0.5, -4]])

p = Polygon(y, facecolor='g', alpha=0.2, edgecolor='k')
p1 = Polygon(y, facecolor='w', edgecolor='k')
p2 = Polygon(y1, facecolor='k', edgecolor='k')

fig, ax = plt.subplots()

ax.add_patch(p1)
ax.add_patch(p)
ax.add_patch(p2)

ax.set_ylim([-4.2, 0.2])
ax.set_xlim([-2.1, 2.1])

plt.axis('off')

plt.show()
```
