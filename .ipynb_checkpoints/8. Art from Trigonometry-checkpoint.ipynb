### Art from Trigonometry

#### 1. Asteroid Radial Curve

```python
t = np.linspace(-6*np.pi, 6*np.pi, 1000)
x = t*np.cos(t)**3
y = t*np.sin(t)**3

plt.style.use('dark_background')

plt.plot(x, y, 'm', linewidth=2)

plt.axis('square')
plt.axis('off')

plt.show()
```

```python
t = np.linspace(-6*np.pi, 6*np.pi, 1000)

fig, ax = plt.subplots(3, 3, figsize=(10, 6))
row, col = np.indices((3, 3))

for i in range(9):
    a = t**i
    x = a * np.cos(t)**3
    y = a * np.sin(t)**3

    r = row.ravel()[i]
    c = col.ravel()[i]
    
    ax[r, c].plot(x,y, 'm')
    ax[r, c].axis('off')
    ax[r, c].axis('square')


plt.show()
```

### 2. Rose Curves

```python
t = np.linspace(0, 8*np.pi, 1000)
k = 4.5
x = np.cos(k*t) * np.cos(t)
y = np.cos(k*t) * np.sin(t)

plt.plot(x, y, 'm', linewidth=2)

plt.axis('square')
plt.axis('off')

plt.show()
```

```python
t = np.linspace(0, 8*np.pi, 1000)

fig, ax = plt.subplots(3, 3, figsize=(10, 6))
row, col = np.indices((3, 3))

for i in range(0, 9):
    k = i/2
    x = np.cos(k*t) * np.cos(t)
    y = np.cos(k*t) * np.sin(t)

    r = row.ravel()[i]
    c = col.ravel()[i]
    
    ax[r, c].plot(x,y, 'm')
    ax[r, c].axis('off')
    ax[r, c].axis('square')
    ax[r, c].set_title('$k = %s$' %k)


plt.show()
```

### 3. Squircle

```python
# Formula for squircle: x**4 + y**4 = a**4

# We isolate y and get y = (1 - x**4)**(1/4)

x = np.linspace(-1, 1, 1000)
y = (1 - x**4)**(1/4)

plt.plot(x, y, 'w')
plt.plot(x, -y, 'w')

plt.axis('square')
plt.axis('off')

plt.show()
```

### 4. Logarithmic spiral

```python
t = np.linspace(0, 10*np.pi, 1000)
k = -3

x = np.cos(t) * np.exp(t/k)
y = np.sin(t) * np.exp(t/k)


plt.plot(x, y, 'm', linewidth=2)
plt.axis('square')
plt.axis('off')

plt.show()
```

```python
n = 150
ks = np.linspace(-6, -2, n)

for i in range(n):
    x = np.cos(t) * np.exp(t/ks[i])
    y = np.sin(t) * np.exp(t/ks[i])
    
    plt.plot(x, y, linewidth=2, color=[i/n*.5, 0, i/n])

plt.axis('square')
plt.axis('off')

plt.show()
```

### 5. Logistic Map

- Created by evaluating this equation: 
$$x_n + 1 = rx_n(1 - x_n)$$
- Where (in our example): 
$$x_1 = 0.5$$
$$n \in \mathbb{N} : n \leq 500$$
$$r \in [1, 4] \; \text{in 1000 steps}$$

```python
import warnings
warnings.filterwarnings('ignore', category=RuntimeWarning)
```

```python
rs = np.linspace(1, 4, 400)
N = 500
x = 0.5 + np.zeros(N)  # makes all values 0.5 which will be replaced later except the first one
endcap = np.arange(int(round(N*0.9)), N)



for ri in range(len(rs)):
    for n in range(N-1):

        x[n+1] = rs[ri]*x[n] * (1-x[n]) + 0.2*np.sin(2*np.pi*x[n])**2
    
    u = np.unique(x[endcap])
    r = rs[ri]*np.ones(len(u))
    plt.plot(r, u, '.', markersize=1, color=[(np.sin(ri/len(rs)/2)+1)/2, 1-ri/len(rs), 0.5])
        

plt.show()
```
